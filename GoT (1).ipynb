```python
# Let's sift through the people involved in attacker kings, defender kings, the types of battle that took place, and their counts. 
```

```python
attacker_king = battles.attacker_king
attacker_king = attacker_king.dropna()
att_king = attacker_king.unique()
print(att_king)
```

```python
count = battles.attacker_king.value_counts()
plt.figure(figsize = (25, 8))
ax2 = sns.barplot(x = count[0:10].index, y = count[0:10], palette = "husl")
plt.xticks(rotation = 360)
plt.title("The number of battles each attacker king pursued")
plt.ylabel("Number of Battles per Attacker King")
print(count[0:10].index)
```
